```python
# Here is an introduction to Pandas, the easy way to structure data
# into tables and extract them from files.

# One of the important components of Pandas is the DataFrame, which
# is essentially a two-dimensional table to store data inside. Let us
# see this in action.

import pandas as pd

# The above line tells Python to download the pandas library, and store
# all the methods and variables that it carries into 'pd'. We will now
# create a string variable that stores the name of the CSV file we will
# be using. A CSV (Comma-seperated values) file contains a data-set.

location = "top_ten_cities_by_pop.csv"

# Now, we will use the read_csv() method of pandas to extract the info
# stored in the CSV. The result will be a DataFrame stored inside 'df'.

df = pd.read_csv(location)
df
```

```python
# Now, as you can see, we have read the complete CSV file. If you want
# to take a small peek at the data, the head() method of DataFrame will
# show the first 5 rows by default.

df.head()
```

```python
# If you want to peek inside a specific number of rows, put that number
# inside the parentheses of head().

df.head(2)

# This also works with the tail() method, where you are taking the last
# few rows.
```
